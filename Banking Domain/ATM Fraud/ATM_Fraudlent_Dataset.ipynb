```python
from sklearn.pipeline import make_pipeline
```

```python
pipelines = []
for model in [RandomForestClassifier(), 
              SVC(),
              KNeighborsClassifier(),
              RadiusNeighborsClassifier()]:
    pipeline = make_pipeline(model)
    pipelines.append(pipeline)
```

```python
param_grid = {}


rf_grid = {'n_estimators': [100, 200, 300],
           'criterion': ['gini', 'entropy']
           }

svc_grid = {'C': [1.0, 3.0, 5.0],
           'kernel': ['linear', 'poly', 'rbf'],
           'degree': [2, 3, 5]
           }

knn_grid = {'n_neighbors': [50, 100, 150],
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
            'metric': ['minkowski'], 
            'p': [1, 2]
           }
rnn_grid = {'radius': [300, 400, 500],
            'weights':['uniform', 'distance'],
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
            'metric': ['minkowski'], 
            'p': [1, 2]
           }
lst = [rf_grid,svc_grid, knn_grid, rnn_grid]
i=0
for pipeline in pipelines:
   
    if(i < len(lst)):
        param_grid[pipeline[0]] = lst[i]
        i = i+1 
```
